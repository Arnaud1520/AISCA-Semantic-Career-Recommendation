# Génération des embeddings sémantiques

Cette étape consiste à transformer le référentiel de compétences construit dans la couche Gold en vecteurs sémantiques.

Ces vecteurs permettront :

- de comparer un profil utilisateur à notre référentiel
- d’évaluer la similarité entre compétences et expérience utilisateur
- d’alimenter le moteur de recommandation métier

Le modèle utilisé est Sentence-BERT (SBERT).

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

# Chargement du référentiel de compétences

```python
competencies_df = pd.read_csv(
    "../data/gold/competencies.csv"
)

competencies_df.head()
```

# Chargement du modèle Sentence-BERT

```python
model = SentenceTransformer(
    "paraphrase-multilingual-MiniLM-L12-v2",
    device="cpu"
)
```

# Extraction des descriptions sémantiques

```python
semantic_texts = competencies_df[
    "semantic_description"
].tolist()

print(semantic_texts)
```

# Génération des vecteurs sémantiques

```python
competency_embeddings = model.encode(
    semantic_texts
)
```

```python
print(competency_embeddings.shape)
```

# Sauvegarde des embeddings

```python
np.save(
    "../data/gold/competency_embeddings.npy",
    competency_embeddings
)
```

```python
loaded_embeddings = np.load(
    "../data/gold/competency_embeddings.npy"
)

print(loaded_embeddings.shape)
```

# Création d’un profil utilisateur fictif

```python
user_profile = """

I have experience building machine learning systems.

I deployed cloud applications using AWS.

I worked with Docker containers.

I developed automation scripts in Python.

"""
```

```python
user_embedding = model.encode(
    user_profile
)
```

```python
print(user_embedding.shape)
```

# Conclusion

Le référentiel de compétences a été transformé en vecteurs sémantiques.

Chaque compétence dispose désormais d’une représentation vectorielle de 384 dimensions.

Ces embeddings seront utilisés dans l’étape suivante afin de calculer la similarité entre le profil utilisateur et les compétences du référentiel.

La prochaine étape consistera à construire le moteur de similarité sémantique.
