# Construction de la couche SILVER

Cette étape consiste à nettoyer les données issues de la couche RAW afin d'obtenir un dataset standardisé et exploitable pour les traitements futurs.

Objectifs :

- Traiter les valeurs manquantes
- Supprimer les données inutilisables
- Standardiser les variables
- Générer un dataset propre destiné à la couche Silver

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/job_market.csv")
```

# Vérification initiale

Observation de l’état initial du dataset avant transformation.

```python
print(df.shape)

df.isnull().sum()
```

# Suppression des lignes sans compétences

La colonne skills représente le référentiel principal utilisé pour l’analyse sémantique.

Les lignes ne contenant aucune compétence ne peuvent donc pas être exploitées.

```python
df = df.dropna(subset=["skills"])
```

```python
print(df.shape)
```

# Traitement des valeurs manquantes dans job_type

Cette variable étant catégorielle, les valeurs manquantes sont remplacées par une valeur par défaut.

```python
df["job_type"] = df["job_type"].fillna("Unknown")
```

```python
print(df["job_type"].isnull().sum())
```

# Traitement des valeurs manquantes dans category

Les catégories absentes sont remplacées par une catégorie générique.

```python
df["category"] = df["category"].fillna("Other")

print(df["category"].isnull().sum())
```

# Traitement de experience_required

Cette variable étant numérique, les valeurs manquantes sont remplacées par la médiane.

```python
median_exp = df["experience_required"].median()

print("Médiane :", median_exp)
```

```python
df["experience_required"] = df["experience_required"].fillna(
    median_exp
)
```

```python
print(df["experience_required"].isnull().sum())
```

# Vérification finale

Nous vérifions que toutes les opérations de nettoyage ont été correctement appliquées.

```python
df.isnull().sum()
```

# Aperçu du dataset nettoyé

```python
df.head()
```

# Export des données nettoyées

Le dataset nettoyé est sauvegardé dans la couche Silver.

```python
df.to_csv("../data/silver/cleaned_jobs.csv", index=False)
```

```python
df_silver = pd.read_csv("../data/silver/cleaned_jobs.csv")

print(df_silver.shape)
```

# Conclusion de la couche SILVER

Les opérations suivantes ont été réalisées :

- Suppression des lignes sans compétences
- Imputation des variables catégorielles
- Imputation des variables numériques
- Standardisation du dataset

Le dataset est maintenant prêt à être transformé en couche Gold.

La couche Gold contiendra les référentiels métiers et compétences utilisés par le moteur de recommandation.
